# <B> Titanic Dataset - Data Cleaning & EDA

### Step 1: Load the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the Titanic dataset

```python
df = pd.read_csv(r"C:\Users\Lenovo\Downloads\Titanic-Dataset.csv")
```

### Display basic information about the dataset

```python
print("Dataset Information:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())
```

# Step 2: Handle Missing Values
## Explanation: The dataset contains missing values in Age, Embarked, and Cabin. We handle them as follows:


```python
# Fill missing Age values with median (since Age is numerical)
df.loc[:, 'Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# Fill missing Embarked values with mode (most frequent category)
df.loc[:, 'Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# Drop Cabin column (too many missing values, making it unreliable)
df.drop(columns=['Cabin'], inplace=True)
```

```python
# Verify missing values are handled
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

## Step 3: Exploratory Data Analysis (EDA)
 Survival Distribution
 Plot showing the distribution of passengers who survived vs. those who didn't

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Survived', data=df, palette='coolwarm', legend=False)
plt.title("Survival Distribution")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

 ### Survival by Gender
 This plot helps us understand whether gender played a role in survival rates.

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=df, palette='coolwarm')
plt.title("Survival by Gender")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

### Survival by Passenger Class
Checking survival rates based on passenger class (Pclass)

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='coolwarm')
plt.title("Survival by Passenger Class")
plt.xlabel("Pclass (1 = First, 2 = Second, 3 = Third)")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

### Age Distribution by Survival
This KDE plot shows how age distribution differs between survivors and non-survivors.

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(x=df['Age'], hue=df['Survived'], fill=True, palette=['red', 'green'], alpha=0.5)
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Density")
plt.legend(title="Survived", labels=["Not Survived", "Survived"])
plt.show()
```

## Step 4: Conclusion



```python
# Summary of insights gained from the EDA
print("The exploratory data analysis of the Titanic dataset reveals that gender and passenger class played a significant role in survival rates. Female passengers had a much higher survival rate compared to males, indicating that women and children were prioritized during rescue efforts. Additionally, first-class passengers had a better chance of survival compared to those in second and third class, likely due to better access to lifeboats and safety measures. The age distribution analysis suggests that younger passengers had a slightly higher survival rate, though the difference was not as pronounced. Overall, these findings highlight key social and logistical factors that influenced survival outcomes on the Titanic.")
```
